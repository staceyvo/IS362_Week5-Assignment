What is the northernmost airport in the United States?

My guess is that the northernmost airport in the United States will be in Alaska, as it is the northernmost state.  Now we'll read and sort the data from .csv file posted on github and display the top 5 airports by latitude.

```python
airports = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
airports.sort_values('lat', ascending=False).head(5)
```

The returned results tell us that Dillant Hopkins Airport is the northernmost based on the latitude of 72.270833.  Upon further inspection of the data, I noticed that the tzone field is inconsistent with the others.  I looked up the coordinates for each airport and discovered that the latitude and longitude coordinates for Dillant Hopkins Airport entered incorrectly.  The true northernmost airport is Wiley Post Will Rogers Memorial Airport.

What is the easternmost airport in the United States?

I believe that the easternmost airport will be in Maine, since no other state is further east.

Now we'll sort the data by longitude in order to figure out the easternmost airport.

```python
airports = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv')
airports[(airports['tzone'].str.contains('America')==True)&(airports['tzone'].str.contains('Anchorage')==False)].sort_values('lon',ascending=[False]).head()
```

Seaching by just longitude values did not return the desired results.  The results received were for Alaska which is in the opposite direction that we are looking for.  In order to find the easternmost aiport, I had to remove the Anchorage Alaska results.  

Doing this returns Eastport Municipal Airport which is located in Maine, and is the easternmost aiport in the United Staes.

On February 12th, 2013, which New York area airport had the windiest weather?



```python
weather = pd.read_csv('https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv')
weather[(weather["month"]==2) & (weather['day']==12) & (weather['year']==2013)].sort_values('wind_speed', ascending=False).head(5)
```
